```python
from pyspark.sql.functions import current_timestamp, input_file_name, col, concat, lit, when, rand, cast, sum 

from pyspark.sql.types import *
```

```python
raw_df = (
    spark.read.table('temp.2_silver.user_dataset')
)

display(raw_df)
```

```python
raw_df = raw_df.groupBy(col("signup_date")).agg(
    sum(col("age")).alias("sum_age")
)
display(raw_df)
```

```python
raw_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("temp.3_gold.user_dataset")
```
